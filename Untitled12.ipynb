```python
import torch # PyTorch ကို သုံးမယ်

# ၁။ Data ပြင်ဆင်ခြင်း
# inputs (X) နဲ့ target (y)
X = torch.tensor([[35.0, 70.0, 15.0]])
y = torch.tensor([[10.0]])

# ၂။ အလွှာတစ်ခုတည်းရှိတဲ့ Neural Network တစ်ခု ဆောက်မယ်
# Linear(3, 1) ဆိုတာ input 3 ခုကနေ output 1 ခု ထုတ်မယ်လို့ ပြောတာပါ
model = torch.nn.Linear(3, 1)

# ၃။ Optimizer (Weights တွေကို ဘယ်သူက လိုက်ပြင်ပေးမှာလဲ?)
# ဒီမှာတော့ 'SGD' ဆိုတဲ့ နည်းလမ်းကို သုံးပြီး learning rate ကို 0.0001 ထားမယ်
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001)

# ၄။ Training Loop (အကြိမ် ၁၀၀ လေ့ကျင့်ပေးမယ်)
for epoch in range(100):
    prediction = model(X) # ရှေ့သို့တွက်ခြင်း (Forward Pass)
    loss = torch.nn.functional.mse_loss(prediction, y) # အမှားရှာခြင်း

    optimizer.zero_grad() # အဟောင်းတွေကို ဖျက်
    loss.backward() # အနောက်သို့ပြန်ပြင်ခြင်း (Backpropagation)
    optimizer.step() # Weights တွေကို အသစ်လဲခြင်း

    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}: အမှား (Loss) = {loss.item():.4f}")
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# ၁။ ပုံတွေကို ဒေါင်းလုဒ်ဆွဲပြီး ပြင်ဆင်ခြင်း
transform = transforms.Compose([transforms.ToTensor()])
trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True)

# ၂။ AI မြင်ရမယ့် ပုံအချို့ကို ထုတ်ကြည့်မယ်
dataiter = iter(trainloader)
images, labels = next(dataiter)

# ပုံလေးတွေကို ပြပေးမယ်
plt.imshow(images[0].numpy().squeeze(), cmap='gray')
plt.title(f"Label: {labels[0]}")
plt.show()
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Convolution Layer: ၅x၅ မှန်ဘီလူးနဲ့ Scan ဖတ်မယ်
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        # Pooling Layer: အရွယ်အစားကို တစ်ဝက်ချုံ့မယ်
        self.pool = nn.MaxPool2d(2, 2)
        # Fully Connected Layer: နောက်ဆုံးမှာ ဘာဂဏန်းလဲဆိုတာ ဆုံးဖြတ်မယ်
        self.fc = nn.Linear(10 * 12 * 12, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x))) # Scan ဖတ်ပြီး ချုံ့မယ်
        x = x.view(-1, 10 * 12 * 12) # ဂဏန်းတွေကို တန်းစီလိုက်မယ်
        x = self.fc(x) # အဖြေထုတ်မယ်
        return x

net = SimpleCNN()
print(net)
```

```python
import torch.optim as optim

# ၁။ အမှားရှာမယ့်နည်းလမ်းနဲ့ Weights ပြင်မယ့်နည်းလမ်း (Optimizer) ကို သတ်မှတ်မယ်
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.9)

# ၂။ Training စတင်မယ် (၂ ကျော့လောက်ပဲ စမ်းသင်ကြည့်မယ်)
for epoch in range(2):
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data
        optimizer.zero_grad() # အဟောင်းတွေကို ဖျက်

        outputs = net(inputs) # ခန့်မှန်းခိုင်းမယ်
        loss = criterion(outputs, labels) # အမှားရှာမယ်
        loss.backward() # အနောက်ပြန်ပြင်မယ်
        optimizer.step() # Weights အသစ်လဲမယ်

        running_loss += loss.item()
        if i % 2000 == 1999: # ၂၀၀၀ တုံးတိုင်းမှာ အခြေအနေ ထုတ်ပြမယ်
            print(f'[{epoch + 1}, {i + 1:5d}] အမှား (loss): {running_loss / 2000:.3f}')
            running_loss = 0.0

print('AI ပညာသင်ယူပြီးပါပြီ!')
```

```python
# ၁။ Test Data အချို့ကို ယူမယ်
testloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True)
dataiter = iter(testloader)
images, labels = next(dataiter)

# ၂။ AI ကို ခန့်မှန်းခိုင်းမယ်
outputs = net(images)
_, predicted = torch.max(outputs, 1)

# ၃။ ရလဒ်ကို ပြမယ်
plt.imshow(images[0].numpy().squeeze(), cmap='gray')
plt.title(f"အမှန်တရား: {labels[0]} | AI ခန့်မှန်းချက်: {predicted[0]}")
plt.show()
```

```python
sentence = "AI is very powerful"
# စာကြောင်းကို အလွယ်ဆုံးနည်းနဲ့ ခွဲကြည့်မယ်
tokens = sentence.split()

print(f"မူရင်းစာကြောင်း: {sentence}")
print(f"ခွဲလိုက်တဲ့ Tokens များ: {tokens}")
```

```python
# စကားလုံးတွေကို ID နံပါတ်ပေးတဲ့ ဇယား (Vocabulary) ဆောက်မယ်
vocab = {"AI": 1, "is": 2, "very": 3, "powerful": 4}

# စာကြောင်းကို ID နံပါတ်တွေအဖြစ် ပြောင်းမယ်
numerical_tokens = [vocab[token] for token in tokens]

print(f"ဂဏန်းအဖြစ် ပြောင်းလဲပြီးသော စာကြောင်း: {numerical_tokens}")
```

```python
import torch
import torch.nn as nn

# စကားလုံး ၄ လုံးအတွက်၊ တစ်လုံးစီကို ဂဏန်း ၅ လုံးပါတဲ့ Vector အဖြစ် ပြောင်းမယ်
embedding = nn.Embedding(5, 5) # (vocab size, embedding dim)
input_ids = torch.tensor(numerical_tokens)

# Embedding ထုတ်ကြည့်မယ်
word_vectors = embedding(input_ids)

print("စကားလုံး တစ်လုံးချင်းစီရဲ့ အဓိပ္ပာယ်မြေပုံ (Embedding Vectors):")
print(word_vectors)
```

```python
# RNN အသေးစားလေး ဆောက်မယ်
# input size 5 (embedding size), hidden size 10 (မှတ်ဉာဏ်အရွယ်အစား)
rnn = nn.RNN(5, 10, batch_first=True)

# ကျွန်တော်တို့ရဲ့ word_vectors ကို RNN ထဲ ထည့်လိုက်မယ်
# batch_size=1 ဖြစ်အောင် shape ပြင်ပေးရပါတယ်
inputs = word_vectors.unsqueeze(0)

output, hidden = rnn(inputs)

print("RNN ရဲ့ နောက်ဆုံး မှတ်ဉာဏ် (Hidden State):")
print(hidden)
```
